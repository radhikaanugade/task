# Install matplotlib and import pyplot as plt

```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Plot a basic line chart using plt.plot()

```python
x=[1,2,3,4,5,6,7]
y=[50,51,52,48,47,49,46]
```

```python
plt.plot(x,y,color='blue',linewidth=5, linestyle='--')
plt.show()
```

# Add title and labels to axes

```python
plt.xlabel('Day')
plt.ylabel('Temperature')
plt.title('Weather')
plt.plot(x,y,color='blue',linewidth=5, linestyle='--')
plt.show()
```

# Change line color,style and markers

```python
plt.plot(x,y,color='red',marker='*',linestyle='')
plt.show()
```

# Plot multiple lines in the same chart

```python
days=[1,2,3,4,5,6,7]
max_t=[50,51,52,48,47,49,46]
min_t=[43,42,40,44,33,35,37]
avg_t=[45,48,48,46,40,42,41]
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.show()
```

# Add a legend tothe chart

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend()
plt.show()
```

# Save plot as an image file

```python
plt.savefig('img,PNG',transparent=None,facecolor='auto', edgecolor='auto')
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.show()
```

# Plot bar chart using plt.bar()

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
company=['GOOGLE','AMZN','MSFT','FB']
revenue=[90,136,89,27]
xpos = np.arange(len(company))
xpos
```

```python
plt.bar(xpos,revenue,label="Revenue")

plt.xticks(xpos,company)
plt.ylabel("Revenue(Bin)")
plt.title('US Technology Stocks')
plt.legend()
plt.show()
```

# Use subplots to display multiple plots

```python
x=[1,2,3,4,5]
y1=[1,4,9,16,25]
y2=[25,16,9,4,1]
```

```python
plt.subplot(1,2,1)
plt.plot(x,y1,'r-o')
plt.title('plot 1')

plt.subplot(1,2,2)
plt.plot(x,y1,'g--s')
plt.title('plot 2')

plt.tight_layout()
plt.show()
```

# Customize ticks and grid lines

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
plt.legend(loc='lower left',shadow=True,fontsize='small')
plt.show()
```

# Use different figure sizes 

```python
plt.figure(figsize=(10,5))
x =[1,2,3,4,5]
y =[10,20,15,25,30]
plt.plot(x,y, 'b-o')
plt.xlabel('x')
plt.ylabel('y')
plt.title('custom figure size')
plt.show()
```

# Create a stacked bar chart

```python
labels = ['A', 'B', 'C', 'D']
group1 = [10, 15, 20, 25]
group2 = [5, 10, 5, 15]


plt.bar(labels, group1, label='Group 1')
plt.bar(labels, group2, bottom=group1, label='Group 2')  
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Stacked Bar Chart')
plt.legend()

plt.show()
```

# Chnage background and axis colors

```python
fig, ax = plt.subplots()
x = [1, 2, 3, 4, 5]
y = [10, 20, 15, 25, 30]
ax.plot(x, y, color='blue')

# Change background color
fig.patch.set_facecolor('lightgray')  
ax.set_facecolor('lavender')              

# Change axis colors
ax.spines['bottom'].set_color('red')
ax.spines['left'].set_color('green')
ax.tick_params(axis='x', colors='red')
ax.tick_params(axis='y', colors='green')

# Title and labels with custom color
ax.set_title('Custom Background and Axis Colors', color='purple')
ax.set_xlabel('X-axis', color='red')
ax.set_ylabel('Y-axis', color='green')
plt.show()
```
