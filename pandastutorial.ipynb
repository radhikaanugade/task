## Delete nan rows

```python
df_new
```

## impute missing values

### sometimes if we delete entire columns thet will be not the appropriate approch. delete column can affect our model building because we will lose our main feature for imputing we have many approches so here are some of the popular technidues

### method 1: impute fixed value like 0,or missing' etc.we impute unknown in gender colimns

```python
df
```

```python
df['Gender'].fillna('Unknown', inplace=True)
df
```

```python
# impute mean in amount_spent columns
mean_amount_spent = df['Amount_spent'].mean()
df['Amount_spent'].fillna(mean_amount_spent, inplace=True)
df
mean_amount_spent
```

```python
# impute median in age column
median_Age = df['Age'].median()
df['Age'].fillna(median_Age, inplace=True)
df
median_Age
```

```python
# impute mode in Employees_status column
mode_Employees_status = df['Employees_status'].mode().iloc[0]
df['Employees_status'].fillna(mode_Employees_status, inplace=True)
mode_Employees_status
```

```python
mean_amount_spent
```

```python
median_Age
```

```python
mode_Employees_status
```

## method 3- imputing forword fill or backfill by ffill and bfill.in ffill missing value imputing from the value of the above row and for bfill its taken from the below rows value

```python
df['Referal'].fillna(method='ffill',inplace=True)
df
```

```python
df.isna().sum().sum()
```

```python
df.isna().sum().sort_values(ascending=False)
```

## 5.3 Summaizing or grouping data

## groupby

Grouping by one column: for example, if we want to find maximum values of age and amount_spent by gendar then we can use this

df[['Age','Amount_spent']].groupby(df['Gender']).max()

to find mean,count and max values of Age an amount_spent by Gender then we can use agg() functions with groupby()

```python
#Groupby one columns
state_gender_res = df[['Age','Gender','Amount_spent']].groupby(['Gender']).agg(['count','mean','max'])
state_gender_res
```

Grouping by multiple columns: to find total count,maximum and minimum values of amount_spent by state_names,gender and payment_method then we can pass these columns name under groupby() functions and add.agg()with  count,mean,max argument


```python
# group by multiple columns
state_genders_res=df[['State_names','Gender','Payment_method','Amount_spent']].groupby(['State_names','Gender','Payment_method']).agg(['count','min','max'])
state_genders_res.head(12)
```
